## AUC

EL UAC (User Acount Control) área bajo la curva ROC, mide el área bidimensioanl completa debajo de tocda curva ROC. AUC proporciona una medida agregada de rendimiento en todos los umbrales de clasificación posibles. Una forma de interpretar el AUC es como la probabilidad de que el modelo clasifique un ejemplo positivo aleatorio más alto que un ejemplo negativo aleatorio.

## CURVA ROC

La curva ROC (Receiver Operating Characteristic) o Caracterísitica Operativa del Receptor, la utilizamos para crear representaciones gráficas especializada para sistemas de clasificación binario.
El análisis de la curva ROC proporciona las herramietas para seleccional el modelo más óptimo sin tener en cuenta el coste de la distribuciónb de las dos clases sobre las que se decide.
